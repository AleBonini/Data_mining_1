```python
%matplotlib inline
import math
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

from collections import defaultdict
from scipy.stats.stats import pearsonr
```

```python
df = pd.read_csv("training.csv")
```

```python
# Set up a grid of plots
fig = plt.figure(figsize=(50, 50)) 
fig_dims = (5, 4)

# Plot Type of transmission
plt.subplot2grid(fig_dims, (0, 0))
df['Transmission'].value_counts().plot(kind='bar', 
                                   title='Type of transmission')

# Plot Vehicle Year
plt.subplot2grid(fig_dims, (0, 1))
df['VehicleAge'].value_counts().plot(kind='bar', title='Vehicle Age')

# Plot TopThreeAmericanName
plt.subplot2grid(fig_dims, (1, 0))
df['TopThreeAmericanName'].value_counts().plot(kind='bar', title='TopThreeAmericanName')
plt.xticks(rotation=0)

# Plot Nationality
plt.subplot2grid(fig_dims, (1, 1))
df['Nationality'].value_counts().plot(kind='bar', title='Nationality')

# Plot Vehicle cost
plt.subplot2grid(fig_dims, (2, 0))
df['VehBCost'].hist(bins = 17)
plt.title('Vehicle cost')

plt.show()
```
